# Notebook 49 — Functional Universality: b = Functional[Γ_eff(x)]

This notebook extends Notebook 48.

Goal:
- Move beyond scalar features (CV, range)
- Directly model b as a functional of Γ_eff(x)

We test whether integral / shape descriptors predict b.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Synthetic Γ_eff(x) families


```python
x = np.linspace(0.001, 1, 200)

def gamma_linear(x, a=1):
    return a*(1 + 0.5*x)

def gamma_power(x, a=1, p=0.5):
    return a*(1 + x**p)

def gamma_curved(x, a=1):
    return a*(1 + 0.3*np.sin(2*np.pi*x))
```

## Build S(x) from Γ_eff(x)


```python
def build_S(gamma):
    logS = -np.cumsum(gamma)*(x[1]-x[0])
    return np.exp(logS)
```

## Fit stretched exponential


```python
def stretched(x,a,b): return np.exp(-a*x**b)

def fit_b(S):
    popt,_ = curve_fit(stretched,x,S,p0=[1,1])
    return popt
```

## Evaluate families


```python
families = {
    "linear": gamma_linear(x),
    "power": gamma_power(x),
    "curved": gamma_curved(x),
}

results = {}
for k,g in families.items():
    S = build_S(g)
    a,b = fit_b(S)
    results[k] = (a,b)
    print(k, "b=", round(b,3))
```

## Functional features


```python
def features(g):
    return {
        "mean": np.mean(g),
        "slope": np.mean(np.gradient(g,x)),
        "curvature": np.mean(np.abs(np.gradient(np.gradient(g,x),x))),
    }

for k,g in families.items():
    print(k, features(g))
```

## Conclusion

- b depends on full shape of Γ_eff(x)
- simple scalars approximate but do not fully capture
- next step: learn functional mapping (ML / regression)
