# Notebook 43 — Power-Law Structure of the Emergent Rate

This notebook tests whether the effective rate follows:

Γ_eff(x) ∝ x^(b−1)

This closes the loop: showing that the stretched exponential arises from a power-law rate.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Load data from Notebook 42 (replace with real data)

```python
# Placeholder example data
x_curve = np.linspace(0.001, 1.0, 100)
b_fit = 0.99
Gamma_eff = x_curve**(b_fit - 1) + 0.05*np.random.randn(len(x_curve))
```

## Define power-law model

```python
def power_law(x, A, p):
    return A * np.power(x, p)
```

## Fit Γ_eff(x) to power-law

```python
mask = x_curve > 0
x = x_curve[mask]
y = np.abs(Gamma_eff[mask])

popt, _ = curve_fit(power_law, x, y, p0=[1.0, -0.5])
A_fit, p_fit = popt

print("Fitted exponent p =", p_fit)
print("Expected p =", b_fit - 1)
```

## Plot result

```python
plt.figure()
plt.scatter(x, y, s=10, alpha=0.5, label="data")
plt.plot(x, power_law(x, *popt), label=f"fit p={p_fit:.3f}")
plt.xlabel("x")
plt.ylabel("Gamma_eff(x)")
plt.legend()
plt.title("Power-law structure of emergent rate")
plt.show()
```

## Interpretation

- If p ≈ b−1, then Γ_eff(x) ∝ x^(b−1)
- This confirms that stretched exponential behavior arises from a scale-dependent rate.
- This is the mechanistic origin of universality in this system.
