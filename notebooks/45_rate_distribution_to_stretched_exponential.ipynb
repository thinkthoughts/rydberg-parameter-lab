# Notebook 45 — From Rate Distribution to Stretched Exponential

Derive how a distribution of decay rates produces the observed stretched-exponential universality.

```

import numpy as np
import matplotlib.pyplot as plt
```

## Define a distribution of decay rates Γ

```

def sample_rates(n=10000, shape=2.0, scale=1.0):
    return np.random.gamma(shape, scale, size=n)

rates = sample_rates()
```

## Compute ensemble decay S(x) = <exp(-Γ x)>

```

x = np.linspace(0, 0.25, 200)

def ensemble_decay(rates, x):
    S = []
    for xi in x:
        S.append(np.mean(np.exp(-rates * xi)))
    return np.array(S)

S = ensemble_decay(rates, x)
```

## Fit stretched exponential

```

from scipy.optimize import curve_fit

def stretched(x, a, b):
    return np.exp(-a * x**b)

params, _ = curve_fit(stretched, x, S, p0=(1, 1))
a_fit, b_fit = params
S_fit = stretched(x, a_fit, b_fit)

print(a_fit, b_fit)
```

## Compare

```

plt.figure(figsize=(6,4))
plt.plot(x, S, label="ensemble decay")
plt.plot(x, S_fit, '--', label=f"stretched fit (b={b_fit:.3f})")
plt.xlabel("x")
plt.ylabel("S")
plt.legend()
plt.title("Emergent stretched exponential from rate distribution")
plt.show()
```

## Insight

A distribution of exponential decays produces a stretched exponential effective law.

This connects Notebook 41–44 results to a physical mechanism.
