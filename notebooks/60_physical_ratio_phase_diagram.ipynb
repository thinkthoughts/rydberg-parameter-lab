# Notebook 60 — Physical Control Ratios → Universality (b)

Map physically meaningful ratios (γ/Ω, V/Ω) → stretched exponent b.

This connects Rydberg control knobs directly to universality classification.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x = np.linspace(1e-3,1,400)

def gamma_eff(x, gamma_over_Omega, V_over_Omega):
    base = 2.0
    deph = gamma_over_Omega * np.exp(-((x-0.3)/0.15)**2)
    interaction = V_over_Omega * np.sin(2*np.pi*x)
    return np.clip(base + deph + interaction, 0.2, None)

def build_S(g):
    dx=x[1]-x[0]
    return np.exp(-np.cumsum(g)*dx)

def stretched(x,a,b): return np.exp(-a*x**b)

def fit_b(S):
    popt,_ = curve_fit(stretched,x,S,p0=[1,1])
    return popt[1]
```

## Sweep physical ratios

```python
gamma_vals = np.linspace(0,1,15)
V_vals = np.linspace(0,0.5,15)

b_map = np.zeros((len(gamma_vals), len(V_vals)))

for i,g in enumerate(gamma_vals):
    for j,v in enumerate(V_vals):
        G = gamma_eff(x,g,v)
        S = build_S(G)
        b_map[i,j] = fit_b(S)
```

## Phase diagram (γ/Ω vs V/Ω)

```python
plt.imshow(b_map, origin='lower', aspect='auto',
           extent=[V_vals[0],V_vals[-1],gamma_vals[0],gamma_vals[-1]])
plt.colorbar(label='b')
plt.xlabel('V / Ω')
plt.ylabel('γ / Ω')
plt.title('Universality phase diagram (physical ratios)')
plt.show()
```

## Interpretation

- Increasing γ/Ω → raises b
- Increasing V/Ω → lowers b
- Universality exponent b is a function of physical control ratios
